### Data Cleaning

```python
# Get first advance per user

# Sort advances dataset by user ID and underwritten date (ascending order)
advances_sorted = advances_df.sort_values(['user_id', 'underwritten_at'])

# Get first record for each user only
first_advances = advances_sorted.groupby('user_id').first().reset_index()

first_advances.head(10)

first_advances.to_csv('data/first_advances.csv', index=False)
```

```python
merged_df = first_advances.merge(labels_df, left_on='request_id', right_on='advance_id', how='inner')
merged_df.head()
```

```python
# Extract list of user IDs that have both first advances and labels available 
user_ids = merged_df['user_id']
user_ids.to_csv('data/user_ids.csv', index=False, header=True)
```

```python
transactions_sorted = transactions_df.sort_values(['user_id', 'date'])
transactions_sorted.head(10)
print(transactions_sorted['primary_category'].nunique())

# print unique values
print(transactions_sorted['primary_category'].unique())
```

```python
transactions_sorted = transactions_df.sort_values(['user_id', 'date'])
transactions_filtered = transactions_sorted[transactions_sorted['user_id'].isin(first_advances['user_id'])]

categories = transactions_sorted['primary_category'].unique().tolist()
categories.remove("TRANSFER_IN") # we are only concerned with outflows
categories.remove("INCOME")

spending = transactions_sorted[(transactions_sorted['primary_category'].isin(categories)) & (transactions_sorted['amount'] > 0)]
spending.groupby(['user_id', 'primary_category'])['amount'].sum().reset_index()

spending_agg = spending.groupby(['user_id', 'primary_category'])['amount'].sum().reset_index()

spending_wide = spending_agg.pivot( 
    index='user_id',
    columns='primary_category',
    values='amount'
).fillna(0)

spending_wide['total_spending'] = spending_wide.sum(axis=1)

category_columns = categories

spending_pct = spending_wide.copy()

for col in category_columns:
    if col in spending_pct.columns:
        # Divide by total to get proportion (0 to 1)
        spending_pct[col] = spending_pct[col] / spending_pct['total_spending']
    else:
        # If category doesn't exist, add it with zeros
        spending_pct[col] = 0

# Reset index to make user_id a column
spending_pct = spending_pct.reset_index()

# Handle users with no spending (all zeros) - fill NaN with 0
spending_pct = spending_pct.fillna(0)


```
