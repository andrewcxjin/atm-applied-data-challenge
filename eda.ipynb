```python
import pandas as pd
import pyarrow.parquet as pq
```

```python
advances = pq.read_table('data/advances.parquet')
advances_df = advances.to_pandas()
advances_df.head()
```

```python
applications = pq.read_table('data/applications.parquet')
applications_df = applications.to_pandas()
applications_df.head()
```

```python
balances = pq.read_table('data/balances.parquet')
balances_df = balances.to_pandas()
balances_df.head()
```

```python
labels = pq.read_table('data/labels.parquet')
labels_df = labels.to_pandas()
labels_df.head()
```

```python
transactions = pq.read_table('data/transactions_sample.parquet')
transactions_df = transactions.to_pandas()
transactions_df.head()
```

```python
dfs = {
    'advances': advances_df,
    'applications': applications_df,
    'balances': balances_df,
    'labels': labels_df,
    'transactions': transactions_df
}
```

```python
for df in dfs.values():
    print(df.size)
```

```python
for name, df in dfs.items():
    print(f'Data Structure of {name}:')
    for column in df.columns:
        print(f'Column: {column}, Type: {df[column].dtype}')
```

```python
for name, df in dfs.items():
    print(f'Null values in {name}:')
    print(df.isnull().mean())
```
