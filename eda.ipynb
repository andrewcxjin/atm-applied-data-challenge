```python
transactions_sorted = transactions_df.sort_values(['user_id', 'date'])
transactions_filtered = transactions_sorted[transactions_sorted['user_id'].isin(first_advances['user_id'])]

categories = transactions_sorted['primary_category'].unique().tolist()
categories.remove("TRANSFER_IN") # we are only concerned with outflows
categories.remove("INCOME")

# Only takes positive amounts as negative represents inflows
spending = transactions_sorted[(transactions_sorted['primary_category'].isin(categories)) & (transactions_sorted['amount'] > 0)]
spending.groupby(['user_id', 'primary_category'])['amount'].sum().reset_index()

spending_agg = spending.groupby(['user_id', 'primary_category'])['amount'].sum().reset_index()

spending_wide = spending_agg.pivot( 
    index='user_id',
    columns='primary_category',
    values='amount'
).fillna(0)

spending_wide['total_spending'] = spending_wide.sum(axis=1)

category_columns = categories

spending_pct = spending_wide.copy()

for col in category_columns:
    if col in spending_pct.columns:
        # Divide by total to get proportion (0 to 1)
        spending_pct[col] = spending_pct[col] / spending_pct['total_spending']
    else:
        # If category doesn't exist, add it with zeros
        spending_pct[col] = 0

# Reset index to make user_id a column
spending_pct = spending_pct.reset_index()

# Handle users with no spending (all zeros) - fill NaN with 0
spending_pct = spending_pct.fillna(0)


```

```python

import numpy as np
from datetime import timedelta


if transactions_sorted['date'].dtype == 'object':
    transactions_sorted = transactions_sorted.copy()
    transactions_sorted['date'] = pd.to_datetime(transactions_sorted['date'])

advance_dates = first_advances.set_index('user_id')['underwritten_at'].apply(pd.to_datetime).to_dict()

user_ids_set = set(first_advances['user_id'].values)
transactions_filtered = transactions_sorted[transactions_sorted['user_id'].isin(user_ids_set)].copy()

transactions_with_advance = transactions_filtered.merge(
    first_advances[['user_id', 'underwritten_at']], 
    on='user_id', 
    how='inner'
)
transactions_with_advance['underwritten_at'] = pd.to_datetime(transactions_with_advance['underwritten_at'])

transactions_before_advance = transactions_with_advance[
    transactions_with_advance['date'] < transactions_with_advance['underwritten_at']
].copy()

print(f"Processing {len(transactions_before_advance)} transactions for {transactions_before_advance['user_id'].nunique()} users")


diversity_all_time = transactions_before_advance.groupby('user_id').agg({
    'merchant_name': lambda x: x.dropna().nunique(),
    'primary_category': 'nunique',
    'detailed_category': 'nunique',
    'date': ['min', 'max', 'count'],
    'id': 'count'
}).reset_index()

diversity_all_time.columns = ['user_id', 'unique_merchants_all_time', 'unique_categories_all_time',
                              'unique_detailed_categories', 'first_transaction_date', 
                              'last_transaction_date', 'date_count', 'total_transactions_all_time']

def calculate_merchant_concentration(group):
    merchant_counts = group['merchant_name'].dropna().value_counts()
    if len(merchant_counts) > 0:
        proportions = merchant_counts / len(merchant_counts)
        return (proportions ** 2).sum()
    return 0

merchant_concentration = transactions_before_advance.groupby('user_id').apply(
    calculate_merchant_concentration
).reset_index(name='merchant_concentration')

def count_recurring_merchants(group):
    merchant_counts = group['merchant_name'].dropna().value_counts()
    return (merchant_counts >= 3).sum()

recurring_merchants = transactions_before_advance.groupby('user_id').apply(
    count_recurring_merchants
).reset_index(name='recurring_merchants_count')

diversity_features_df = diversity_all_time.merge(merchant_concentration, on='user_id', how='left')
diversity_features_df = diversity_features_df.merge(recurring_merchants, on='user_id', how='left')

diversity_features_df['recurring_merchant_pct'] = (
    diversity_features_df['recurring_merchants_count'] / 
    diversity_features_df['unique_merchants_all_time'].replace(0, 1)
)

diversity_features_df['time_span_days'] = (
    (diversity_features_df['last_transaction_date'] - diversity_features_df['first_transaction_date']).dt.days
)
diversity_features_df['transaction_frequency_all_time'] = (
    diversity_features_df['total_transactions_all_time'] / diversity_features_df['time_span_days'].replace(0, 1)
)

diversity_features_df = diversity_features_df.merge(
    first_advances[['user_id', 'underwritten_at']], 
    on='user_id', 
    how='left'
)
diversity_features_df['underwritten_at'] = pd.to_datetime(diversity_features_df['underwritten_at'])
diversity_features_df['days_since_last_transaction'] = (
    (diversity_features_df['underwritten_at'] - diversity_features_df['last_transaction_date']).dt.days
)

for lookback_days in [30, 60, 90]:
    transactions_with_advance[f'cutoff_{lookback_days}d'] = (
        transactions_with_advance['underwritten_at'] - timedelta(days=lookback_days)
    )
    window_transactions = transactions_with_advance[
        (transactions_with_advance['date'] >= transactions_with_advance[f'cutoff_{lookback_days}d']) &
        (transactions_with_advance['date'] < transactions_with_advance['underwritten_at'])
    ].copy()
    
    window_diversity = window_transactions.groupby('user_id').agg({
        'merchant_name': lambda x: x.dropna().nunique(),
        'primary_category': 'nunique',
        'id': 'count'
    }).reset_index()
    
    window_diversity.columns = ['user_id', f'unique_merchants_{lookback_days}d',
                                f'unique_categories_{lookback_days}d', f'transaction_count_{lookback_days}d']
    
    diversity_features_df = diversity_features_df.merge(window_diversity, on='user_id', how='left')

all_users = first_advances[['user_id']].copy()
diversity_features_df = all_users.merge(diversity_features_df, on='user_id', how='left')

cols_to_fill_zero = ['unique_merchants_all_time', 'unique_categories_all_time', 
                     'unique_detailed_categories', 'merchant_concentration',
                     'recurring_merchants_count', 'recurring_merchant_pct',
                     'transaction_frequency_all_time', 'total_transactions_all_time',
                     'unique_merchants_30d', 'unique_categories_30d', 'transaction_count_30d',
                     'unique_merchants_60d', 'unique_categories_60d', 'transaction_count_60d',
                     'unique_merchants_90d', 'unique_categories_90d', 'transaction_count_90d']

for col in cols_to_fill_zero:
    if col in diversity_features_df.columns:
        diversity_features_df[col] = diversity_features_df[col].fillna(0)

if diversity_features_df['days_since_last_transaction'].notna().any():
    max_days = diversity_features_df['days_since_last_transaction'].max()
    diversity_features_df['days_since_last_transaction'] = diversity_features_df['days_since_last_transaction'].fillna(max_days)
else:
    diversity_features_df['days_since_last_transaction'] = diversity_features_df['days_since_last_transaction'].fillna(0)

diversity_features_df = diversity_features_df.drop(columns=[
    'first_transaction_date', 'last_transaction_date', 'date_count', 
    'time_span_days', 'underwritten_at'
], errors='ignore')

print("Transaction Behavior Diversity Features Summary:")
print(diversity_features_df.head(10))
print(f"\nTotal users: {len(diversity_features_df)}")
print(f"\nDiversity Features Statistics:")
print(diversity_features_df[['unique_merchants_all_time', 'unique_categories_all_time', 
                              'merchant_concentration', 'days_since_last_transaction',
                              'recurring_merchants_count', 'transaction_frequency_all_time']].describe())
```

```python
inflow_categories = ['INCOME', 'TRANSFER_IN']
inflow_transactions = transactions_sorted[
    (transactions_sorted['primary_category'].isin(inflow_categories)) & 
    (transactions_sorted['amount'] < 0)
]

inflow_agg = inflow_transactions.groupby('user_id')['amount'].sum().abs().reset_index()
inflow_agg.columns = ['user_id', 'total_inflow']

income_only = transactions_sorted[
    (transactions_sorted['primary_category'] == 'INCOME') & 
    (transactions_sorted['amount'] < 0)
]
income_agg = income_only.groupby('user_id')['amount'].sum().abs().reset_index()
income_agg.columns = ['user_id', 'total_income']

transfer_in_only = transactions_sorted[
    (transactions_sorted['primary_category'] == 'TRANSFER_IN') & 
    (transactions_sorted['amount'] < 0)
]
transfer_in_agg = transfer_in_only.groupby('user_id')['amount'].sum().abs().reset_index()
transfer_in_agg.columns = ['user_id', 'total_transfer_in']

outflow_agg = spending_wide[['total_spending']].reset_index()
outflow_agg.columns = ['user_id', 'total_outflow']

flow_metrics = inflow_agg.merge(income_agg, on='user_id', how='outer').fillna(0)
flow_metrics = flow_metrics.merge(transfer_in_agg, on='user_id', how='outer').fillna(0)
flow_metrics = flow_metrics.merge(outflow_agg, on='user_id', how='outer').fillna(0)

flow_metrics['outflow_inflow_ratio'] = flow_metrics['total_outflow'] / flow_metrics['total_inflow'].replace(0, 1)

flow_metrics['income_spent_pct'] = flow_metrics['total_outflow'] / flow_metrics['total_income'].replace(0, 1)
flow_metrics['income_spent_pct'] = flow_metrics['income_spent_pct'].clip(0, 10)

print("Flow Metrics Summary:")
print(flow_metrics.head(20))
print(f"\nTotal users in flow_metrics: {len(flow_metrics)}")
```
